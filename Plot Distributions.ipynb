```python
import numpy as np
def ecdf(data):
    # compute ECDF
    x = np.sort(data)
    n = x.size
    np.arange(1, n+1)
    y = np.arange(1, n+1) / n
    return (x,y)
```

```python
import numpy as np
import pandas as pd
import plotly.graph_objects as go

def plot_histogram(filename, asn):
    asn_name =getNetworkName(asn)
    label = asn_name + " Round Trip Times"
    
    data = pd.read_csv(filename) 
    #print(data)
    df = pd.DataFrame(data, columns=['RTT'])
    x = df["RTT"].to_numpy()
    
    fig = go.Figure(data=[go.Histogram(x=x, 
                                       xbins=dict(
                                        start=0,
                                        end=1000,
                                        size=10),
                                       autobinx=False)])
    fig.update_layout(
        title_text=label, # title of plot
        xaxis_title_text='milliseconds', # xaxis label
        yaxis_title_text='Count', # yaxis label
        bargap=0.2, # gap between bars of adjacent location coordinates
        bargroupgap=0.1 # gap between bars of the same location coordinates
    )

    fig.show()
```

```python
import os
import sys
# Get the files for the historgrams, CDFs, and PDFs
directory = "/home/ubuntu/NetworkDashboard"
for filename in os.listdir(directory):
    if filename.__contains__("nov_raw") and filename.endswith(".csv"):
        print(filename)
        fn_list1 = filename.split(".")
        print(fn_list1)
        fn1 = fn_list1[0]
        fn2 = fn1.split("_")
        print(fn2)
        network = int(fn2[3])
        plot_histogram(filename,network)
```
